```python
import numpy as np
import pandas as pd

print(f"Starting shape before cleaning: {df_trimmed.shape}\n")

# =====================================================================
# STEP 1: SCOPE FILTERING (NAICS 54xxxx)
# =====================================================================
# Ensure naics_code is handled as a string to allow robust prefix matching
df_trimmed['naics_code'] = df_trimmed['naics_code'].astype(str).str.strip()

# Filter to Professional, Scientific, and Technical Services (Starts with 54)
df_scoped = df_trimmed[df_trimmed['naics_code'].str.startswith('54')].copy()
naics_dropped = len(df_trimmed) - len(df_scoped)

# =====================================================================
# STEP 2: REMOVE DUPLICATES
# =====================================================================
# For a transaction-level dataset, the unique business key is the transaction key.
initial_count = len(df_scoped)
df_scoped = df_scoped.drop_duplicates(subset=['contract_transaction_unique_key'], keep='first')
duplicates_dropped = initial_count - len(df_scoped)

# =====================================================================
# STEP 3: SENSIBLE MISSING VALUE HANDLING (IMPUTATION)
# =====================================================================
# Text Fields: Fill NaN with empty strings so NLP vectorizers don't crash
text_cols = [
    'prime_award_base_transaction_description', 'transaction_description', 
    'product_or_service_code_description', 'naics_description', 
    'parent_award_id_piid', 'solicitation_identifier', 'award_type',
    'type_of_set_aside', 'usaspending_permalink'
]
for col in text_cols:
    if col in df_scoped.columns:
        df_scoped[col] = df_scoped[col].fillna('').astype(str)

# Financial Fields: Missing or null currency values are treated as $0.00
financial_cols = [
    'federal_action_obligation', 'total_dollars_obligated', 
    'current_total_value_of_award', 'base_and_all_options_value'
]
for col in financial_cols:
    if col in df_scoped.columns:
        df_scoped[col] = df_scoped[col].fillna(0.0).astype(float)

# Categorical Metadata: Impute structured missing fields
if 'number_of_offers_received' in df_scoped.columns:
    df_scoped['number_of_offers_received'] = df_scoped['number_of_offers_received'].fillna(-1).astype(int)

# =====================================================================
# STEP 4: DATA TRANSFORMATIONS & FEATURE ENGINEERING
# =====================================================================
# Transformation A: Datetime conversions & strict NaT handling
date_cols = ['action_date', 'period_of_performance_start_date', 'period_of_performance_potential_end_date', 'last_modified_date']
for col in date_cols:
    if col in df_scoped.columns:
        # Convert to datetime to handle validation, coerce broken strings to NaT
        df_scoped[col] = pd.to_datetime(df_scoped[col], errors='coerce')
        # Fill missing dates with a standardized historical placeholder date
        df_scoped[col] = df_scoped[col].fillna(pd.Timestamp('1970-01-01'))
        # Cast back to string so pandas doesn't revive NaT tokens
        df_scoped[col] = df_scoped[col].dt.strftime('%Y-%m-%d')

# Transformation B: Engineered NLP Target Block (Combines base scope and granularity)
df_scoped['nlp_target_text'] = (
    df_scoped['prime_award_base_transaction_description'].str.upper() + " | " + 
    df_scoped['product_or_service_code_description'].str.upper()
).str.strip()

# Transformation C: Engineered Financial Burn Metric
df_scoped['funding_utilization_rate'] = np.where(
    df_scoped['current_total_value_of_award'] > 0,
    (df_scoped['total_dollars_obligated'] / df_scoped['current_total_value_of_award']).round(4),
    0.0
)

# Catch-all safety step to clear out any unmapped anomalies
df_scoped = df_scoped.fillna('')

# =====================================================================
# STEP 5: EXPORT AND AUDIT METRICS LOGGING
# =====================================================================
df_scoped.to_csv('ey_fy2026_cleaned_scoped.csv', index=False)

print("="*55)
print("     PIPELINE CLEANING & TRANSFORMATION RECORD")
print("="*55)
print(f"1. Scope Filter        : Dropped {naics_dropped} rows outside of NAICS 54xxx.")
print(f"2. Deduplication       : Removed {duplicates_dropped} duplicate transaction records.")
print(f"3. Imputation          : Standardized text NaNs to '', financials to 0.0.")
print(f"4. Timeline Normalization: Imputed blank date fields to '1970-01-01'.")
print(f"5. NLP Engineering     : Created composite 'nlp_target_text' matrix input.")
print(f"6. Financial Evaluation: Generated 'funding_utilization_rate' profile feature.")
print("-"*55)
print(f"Final Cleaned Dataset Shape: {df_scoped.shape}")
print(f"Remaining Missing Cells    : {df_scoped.isnull().sum().sum()} (Target: 0)")
print("="*55 + "\n🚀 Cleaned dataset successfully compiled and saved!")
```

```python
import numpy as np
import pandas as pd
import re

# Load our cleaned workspace file
df_transformed = pd.read_csv('ey_fy2026_cleaned_scoped.csv')

# =====================================================================
# TRANSFORMATION 1: NLP TEXT TEXT STANDARDIZATION & CLEANING
# =====================================================================
def clean_nlp_text(text):
    if not text or text == 'nan':
        return ""
    text = text.lower()
    # Remove fiscal year references (e.g., fy24, fy2026)
    text = re.sub(r'\bfy\d{2,4}\b', '', text)
    # Remove common regulatory codes or administrative markers
    text = re.sub(r'\ba-\d+\b', '', text)
    # Strip out standalone punctuation or structural pipe artifacts
    text = re.sub(r'[^a-z\s]', ' ', text)
    # Minimize whitespace bloat
    return " ".join(text.split())

df_transformed['nlp_clean_text'] = df_transformed['nlp_target_text'].astype(str).apply(clean_nlp_text)

# =====================================================================
# TRANSFORMATION 2: LOGARITHMIC SCALE FOR VARIANCE STABILIZATION
# =====================================================================
# Filter out negative values (de-obligations) to handle log bounds smoothly
df_transformed['log_current_award_value'] = np.log10(df_transformed['current_total_value_of_award'].clip(lower=0) + 1)
df_transformed['log_potential_ceiling_value'] = np.log10(df_transformed['base_and_all_options_value'].clip(lower=0) + 1)

# =====================================================================
# TRANSFORMATION 3: TEMPORAL DELTAS & RECOMPETE WINDOWS
# =====================================================================
# Re-enforce datetime format inside the transformation layer
start_date = pd.to_datetime(df_transformed['period_of_performance_start_date'], errors='coerce')
end_date = pd.to_datetime(df_transformed['period_of_performance_potential_end_date'], errors='coerce')

# Fallback today's date context (June 2026 baseline window)
current_pipeline_date = pd.Timestamp('2026-06-23')

# Calculate durations in integer format
df_transformed['contract_duration_days'] = (end_date - start_date).dt.days.fillna(0).astype(int)
df_transformed['days_until_recompete'] = (end_date - current_pipeline_date).dt.days.fillna(0).astype(int)

# =====================================================================
# EXPORT FINAL PIPELINE RUNTIME
# =====================================================================
df_transformed = df_transformed.fillna('')
df_transformed.to_csv('ey_fy2026_transformed_final.csv', index=False)

print("=== Applied Analytical Transformations ===")
print("✅ Created 'nlp_clean_text' (Stripped of codes, lowercase, token-ready).")
print("✅ Engineered 'log_current_award_value' & 'log_potential_ceiling_value'.")
print("✅ Calculated 'contract_duration_days' and relative 'days_until_recompete' horizons.")
print(f"Final Data Matrix Export Shape: {df_transformed.shape}")
print("=========================================\n🚀 Dataset is 100% ready for vector embeddings!")
print(f"Missing vals: {df_transformed.isnull().sum().sum()}")
```
